# DistilBERT para detección de hate speech

Notebook de integración del modelo transformer entrenado en Kaggle.
El modelo fue entrenado con GPU T4 en Kaggle y publicado en Hugging Face Hub
para su uso en producción sin necesidad de incluir los pesos en el repositorio.

Comparativa directa con el modelo clásico:
| Métrica        | Random Forest | DistilBERT |
|----------------|---------------|------------|
| Accuracy       | 59.5%         | 76%        |
| Recall tóxicos | 17%           | 71%        |
| F1 tóxicos     | 0.28          | 0.73       |

El transformer mejora significativamente gracias a su capacidad
de entender el contexto completo de una frase, resolviendo la limitación
principal del modelo clásico basado en bolsa de palabras.
